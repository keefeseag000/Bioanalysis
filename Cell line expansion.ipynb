```python
def fn0_raw_to_dataframe():
    
    """
    Raw data prep function:
    
    dependancy - dictionary with the title of data_dictionary. Must contain Keys: "datetime", "passage/flask", "culture_day"
        "VCD", "viability".
        
    dictionary values must be entered with the following format:
    "datetime": [201802051155,201802051750]  # year-month-day-hour-minute
    "passage/flask#":["p4d","p5d", "p6d", "p2a", "p3a"]    # p4d = passage 4, flask d  
    "culture_day":["4", "5"]
    "VCD":[9.62, 5.02]
    "viability":[98.9, 98.8]
    
    returns a pandas dateframe with datetime index, and new column "flask_id" - last character from "passage/flask#"
    """
    
    
    #checking the length of each dictionary entry
    
    column_a = len(data_dictionary["datetime"])
    column_b = len(data_dictionary["passage/flask#"])
    column_c = len(data_dictionary["culture_day"])
    column_d = len(data_dictionary["VCD"])
    column_e = len(data_dictionary["viability"])
    
    if column_a == column_b == column_c == column_d == column_e:
        print("All columns are the same length of: " + str(len(data_dictionary["datetime"])))

    else:
        print("go back and check your data entry")
        print("datetime has " + str(len(data_dictionary["datetime"])) + " entries")
        print("passage/flask# has " + str(len(data_dictionary["passage/flask#"])) + " entries")
        print("culture_day has " + str(len(data_dictionary["culture_day"])) + " entries")
        print("VCD has " + str(len(data_dictionary["VCD"])) + " entries")
        print("Viability has " + str(len(data_dictionary["viability"])) + " entries")
        return 
    

    #convert datetime column to str, strings are needed for datetime index
    data_dictionary["datetime"] = [str(i) for i in data_dictionary["datetime"]]

    #creating a gloval dataframe 
    global df0_raw_data
    df0_raw_data = pd.DataFrame(data_dictionary)
    
    #create index
    date_index = pd.DatetimeIndex(df0_raw_data["datetime"])
    
    #set index 
    df0_raw_data.set_index(date_index, inplace = True)
    
    #column for flask number - extracting last character from "flask id"
    df0_raw_data["flask_id"] = [i[len(i)-1] for i in df0_raw_data["passage/flask#"]]
    
    return df0_raw_data
```

```python
def extract_unique_flasks():
    """
    Creates a global dictionary
    
    Will create a dictionary with as many dataframes as unique flask ids.
    
    Key = "df_flask_#"  where # is flask id
    value = pandas dataframe filtered for that flask id
    
    Will return the said dictionary. 
    
    """
    the_list = list(df0_raw_data.flask_id.unique()) #unique values if "flask_id" column
    global my_dict
    my_dict = {} 
    for i in the_list:
        my_dict["df_flask_" + i] = df0_raw_data[df0_raw_data["flask_id"] == i]
    return my_dict
```

### Creating a column in each dataframe which counts elapsed time

```python
def Time_count_column():

    """
    Appends a Time_diff column to each pandas dataframe in a dictionary my_dict.
    Calculates the amount of time elapsed (days) between each row datetime index in the dataframe. 
    
    Dependancy: my_dict global dictionary with pandas dataframes for values
    """
    for n in list(my_dict.keys()):
    
        dataframe = my_dict[n] #extracting nth dataframe 
        dataframe.loc[:,"Time_diff"] = dataframe.index.to_series().diff() #adding a "Time_diff" column: time diff between rows
    
        #Setting the new column to a float object and in units of days (by dividing by np.timedelta64(1,"D)
        dataframe.loc[:,"Time_diff"] = dataframe.loc[:,"Time_diff"]/np.timedelta64(1,"D")
    
        #setting first time point as zero (from NaN)
        dataframe.loc[str(dataframe.index[0]),"Time_diff"] = 0  
    
        my_dict[n] = dataframe #placing replacing the new dataframe into the dictionary 
        
    return
```

### Function that counts runtime per dataframe dictionary - need to add description to the function

```python
def Count_time_diff():
    
    """
    Converts the column Time_diff in every dataframe in global my_dict from time difference between rows
    into elapsted time from first row. 
    """

    for n in list(my_dict.keys()): #for every dataframe in the dictionary 
        dataframe = my_dict[n] 
    
        column = dataframe.columns.get_loc("Time_diff") #location of the Time_diff column
        row = 0 
    
        for i in dataframe.loc[:,"Time_diff"]: 
        
            if row == 0:  #skip the first row
                row += 1
                continue 
            else:  #adding the previous row value to the current row value in Time_diff column
                dataframe.iloc[row,column] = dataframe.iloc[(row-1), column] + dataframe.iloc[row, column]
                row += 1 
        my_dict[n] = dataframe
        
    return 
```

### Creating column to count doubling time between each set of two points

```python
def Doubling_time():
    
    """
    Computes doubling time for every dataframe in global my_dict from columns: Time_diff and VCD
    
    """
    
    for n in list(my_dict.keys()):
        
        df_ = my_dict[n]
        df_.loc[:,"Doubling_time"] = np.nan #create a new column and fill with Nan

        row = 0
        column = df_.columns.get_loc("Doubling_time") 

        for i in range(len(df_["Doubling_time"])): #how many rows in this column

            if row == 0:  #ignore first row 
                row += 1
                continue
    
            else:
                a = np.log(df_.iloc[i, df_.columns.get_loc("VCD")]) # natural log of ith row
                b = np.log(df_.iloc[i-1, df_.columns.get_loc("VCD")]) # natural log of ith - 1 row 
                time_a = df_.iloc[i, df_.columns.get_loc("Time_diff")]
                time_b = df_.iloc[i-1, df_.columns.get_loc("Time_diff")]
        
                if a < b:  #fill with Nan if there is no growth
                    df_.iloc[i,column] = np.nan
                    row +=1
                else: 
                    mu = (a-b)/(time_a-time_b)
                    td = (np.log(2)/mu)*24
                    df_.iloc[i,column] = td
                    row += 1
                    
                    
        my_dict[n] = df_
```

```python
def Generations():
    
    """
    Computes the generations of every dataframe in global my_dict from columns: Time_diff and VCD
    """

    for n in list(my_dict.keys()): #iterating through each dataframe in dictionary

        df_ = my_dict[n] 
        df_.loc[:,"Generations"] = np.nan #create a new column and fill with Nan

        row = 0  #starting row reference 
        column = df_.columns.get_loc("Generations") #keep track of column position

        generation = 0 #function variable to add previous generation value 

        for i in range(len(df_["Generations"])): #iterate over number of rows

            if row == 0:  #ignore first row 
                row += 1
                continue
    
            else:
                a = np.log2(df_.iloc[i, df_.columns.get_loc("VCD")]) # natural log of ith row
                b = np.log2(df_.iloc[i-1, df_.columns.get_loc("VCD")]) # natural log of ith - 1 row 
        
                if a < b:  #fill with Nan if there is no growth
                    df_.iloc[i,column] = np.nan
                    #print("Nan")
                    row +=1
                else: 
                    log_ratio = (a-b)
                    generation += log_ratio
            
                    df_.iloc[i,column] = generation 
                    row += 1
                    
        my_dict[n] = df_

            #print(generation)
```

### Visualizing all the dataframe in 4-pane plot 

```python
def global_color():
    """
    A function that creates global dictionary with color keys and rgba values to be used
    in plotting. 
    """
    
    grey = matplotlib.cm.get_cmap("Greys")
    purple = matplotlib.cm.get_cmap("Purples")
    blue = matplotlib.cm.get_cmap("Blues")
    green = matplotlib.cm.get_cmap("Greens")
    orange = matplotlib.cm.get_cmap("Oranges")
    red = matplotlib.cm.get_cmap("Reds")

    cmap_dict = {"grey":grey,
                 "purple":purple,
                 "blue":blue,
                 "green":green,
                 "orange":orange,
                 "red":red}

    global color_dict 
    
    color_dict = {}

    for key, value in cmap_dict.items():
        for i in list(range(1,11)):
            color_dict[key+str(i)] = value(i/10)
```

```python
def dictionary_func():
    
    """
    User input color for every dataframe in the global my_dict. This color will be plotted for that dataset
    using additional functions. 
    """
    
    n = len(list(my_dict.keys()))
    

    global user_colors
    user_colors = {}
    
    print("Number of datasets: " + str(n) )
    print("Dataset names: " + str(list(my_dict.keys())))
    print("""Enter the color and shade (integer 1-10) for each dataset. The options are:
            grey, purple, blue, green, orange, red. For example: red1 = light red 
            and blue10 = dark blue.
    """)
    
    
    
    #Color inputs for dict keys
    for n in list(my_dict.keys()):
        value = input("Choose color for " + str(n) +" : ")
        user_colors[n] = color_dict[value] #accessing global master color_dict
        
    return 
    #color_dict = dict.fromkeys(my_dict.keys()) #initializing dict with keys from arg and None as values 
    
    
```

```python
def expansion_plot():

    x = 1.5 #figsize percent shrink
    font = {"fontname":"calibri"}
    run_lngth = 24 #x-axis length

    fig = plt.figure(figsize = (15.0*x,10.0*x))
    fig.subplots_adjust(hspace=0.35,wspace=0.15) #spacing between subplots

    ax_VCD = fig.add_subplot(221)
    ax_VCD.set(xlim=[0,run_lngth], ylim=[0,12])
    ax_VCD.set_ylabel("VCD 10E6 cells/ml", **font, size = 20, weight = "bold")
    ax_VCD.set_title("Viable Cell Density", size = 26, **font, weight = "bold")
    for key, df in my_dict.items():
        ax_VCD.plot(df["Time_diff"],df["VCD"], marker = "o" , color = user_colors[key], label = key )

    ax_via = fig.add_subplot(222)
    ax_via.set(xlim=[0,run_lngth], ylim=[80,100])
    ax_via.set_ylabel("Viable %", **font, size = 20, weight = "bold")
    ax_via.set_title("Viability", size = 26, **font, weight = "bold")
    for key, df in my_dict.items():
        ax_via.plot(df["Time_diff"],df["viability"], marker = "o" , color = user_colors[key], label = key )

    ax_td = fig.add_subplot(223)
    ax_td.set(xlim=[0,run_lngth], ylim=[10,40])
    ax_td.set_ylabel("hours", **font, size = 20, weight = "bold")
    ax_td.set_title("Doubling Time", size = 26, **font, weight = "bold")
    for key, df in my_dict.items():
        ax_td.plot(df["Time_diff"],df["Doubling_time"], marker = "o", color = user_colors[key], 
                   label = key, linewidth = 0 )

    ax_gen = fig.add_subplot(224)
    ax_gen.set(xlim=[0,run_lngth], ylim=[0,35])
    ax_gen.set_ylabel("Generations", **font, size = 20, weight = "bold")
    ax_gen.set_title("Generations", size = 26, **font, weight = "bold")
    for key, df in my_dict.items():
        ax_gen.plot(df["Time_diff"],df["Generations"], marker = "o", color = user_colors[key], 
                    label = key, linewidth = 0  )


    all_axes = [ax_VCD,ax_via,ax_td,ax_gen]

    for ax in all_axes:
        ax.set_xlabel("Run Time (days)", **font, size = 20, weight = "bold",)

        # shrink current (axes or subplot) by 20%
        box = ax.get_position()
        ax.set_position([box.x0, box.y0, box.width *0.85, box.height])

        #right edge invisible, grey color the rest
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_color("grey")
        ax.spines["top"].set_color("grey")
        ax.spines["bottom"].set_color("grey")


        #removing the axis tick marks
        ax.tick_params(axis = "both", bottom = False, left = False)

        #larger tick labels 
        ax.tick_params(axis="both", which = "major", labelsize = 14)

        #moving x-axis label downward 
        ax.xaxis.labelpad = 15

        #horizontal lines along major tick marks
        ax.yaxis.grid(True, color = "grey" )

        # put a legend to the right of the current (axes/subplot?)
        leg = ax.legend(loc = "center left", bbox_to_anchor = (1, 0.5), labelspacing = 6, frameon = False,)

    #setting the legend line thickness for the whole plot
    leg_lines = leg.get_lines() 
    plt.setp(leg_lines, linewidth = 2)

    plt.savefig("plot.png", dpi = 600)
    
    return
```
