```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_results = pd.read_csv("Rebel Results.csv")
```

```python
df_results = df_results[6:]
```

```python
df_R0015 = df_results[df_results["Sample"] == "R0015"]
```

```python
df_R0015.columns
```

```python
columns = ['Ala',
       'AQ', 'Arg', 'Asn', 'Asp', 'B1', 'B6-OH', 'B6-Oxo', 'Choline', 'Cit',
       'Cystine', 'GABA', 'Gln', 'Glu', 'Glx', 'Gly', 'His', 'Hyp', 'Ile',
       'Leu', 'Lys', 'Met', 'NAM', 'Phe', 'Pro', 'Sarcosine', 'Ser', 'Thr',
       'Trp', 'Tyr', 'Val', '?Ala']
df_R0015.reindex(columns, axis =1)
```

```python
df_results
```

```python
vari = {"R0015": "Perfusion, high VCD", "R0016":"Perfusion, 50kDa",
       "R0017":"Fed-Batch, control"}
clms_list = ['Ala',
       'AQ', 'Arg', 'Asn', 'Asp', 'B1', 'B6-OH', 'B6-Oxo', 'Choline', 'Cit',
       'Cystine', 'GABA', 'Gln', 'Glu', 'Glx', 'Gly', 'His', 'Hyp', 'Ile',
       'Leu', 'Lys', 'Met', 'NAM', 'Phe', 'Pro', 'Sarcosine', 'Ser', 'Thr',
       'Trp', 'Tyr', 'Val', '?Ala']

fig = plt.figure(figsize=(35, 25))

num = 1
for i in clms_list:
    # creating a column of 3 subplots
    ax = fig.add_subplot(6, 6, num)
    num += 1

    # iterating over grouped reactor ID
    for key, grp in df_results.groupby(['Sample']):
        ax.scatter(grp['Reactor Day'], grp[i], label='_nolegend_')  # Point plots
        mask = np.isfinite(grp[i])
        ax.plot(grp['Reactor Day'][mask], grp[i][mask], label=(key + " " + vari[key]))

    ymin, ymax = ax.get_ylim()  # get the min and max of respective axes
    ax.set_ylim(bottom=0 , top=ymax * 1.05) 
    
    

    ax.set_ylabel(i,fontsize=25) 
    
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor=(0.81, .644), loc="upper left", fontsize=25)
fig.tight_layout()
fig.subplots_adjust(right=0.81)
plt.savefig("Rebel results" + ".png", dpi=500)

    
```
